```python
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder \
    .appName("CSV to Parquet Conversion") \
    .getOrCreate()
```

### **Read CSV file**
Uploading the csv file manually to Lakehouse and converting it to delta table as shown below.

```python

# Read the CSV file
csv_file_path = "abfss://377d6253-22a1-4a71-9e02-ce043c0187d0@onelake.dfs.fabric.microsoft.com/5128015b-75b7-4dc6-9e82-05705a301a66/Files/HotelReviewsData/Hotel_Reviews.csv"  # Replace with your CSV file path
df = spark.read.csv(csv_file_path, header=True, inferSchema=True)

display(df)
```

### **Write Data to to Delta table**

```python
df.write.format("delta").saveAsTable("HotelReviews")
```
